```python
df_focus_all=pd.DataFrame()
for comid in comid_list:
    for i in range(length):
        print(comid[i])
        catch_char_url1 = ('https://ceamdev.ceeopdev.net/hms/rest/api/info/catchment?comid=' + str(comid[i]))
        print(catch_char_url1)
        response2 = requests.get(catch_char_url1)
        catch_char1 = response2.json()
        #print(catch_char1)
        df_test = pd.DataFrame(output1['data'][comid[i]]['StreamHydrology']['data'].items(), columns=['Date', 'Strflw'])
       # df_test = pd.DataFrame(columns={'COMID','StrSlope','StrDepth', 'StrXArea'})
        #df_test['Strflw'].replace('[', '').replace(']', '')
        df_test['COMID'] = comid[i]
        df_test['StrSlope'] = catch_char1['metadata']['SLOPE']
        df_test['StrDepth'] = catch_char1['metadata']['BANKFULL_DEPTH']
        df_test['StrXArea'] = catch_char1['metadata']['BANKFULL_XSEC_AREA']
        #flow_q =[]
        print(df_test)
        df_focus = pd.DataFrame(data= df_test.loc[df_test['Date']=='07/01/2014'], columns={'COMID','StrSlope','StrDepth', 'StrXArea'})
        print(df_focus.loc[0, :])
        #df_focus_2 = df_focus.loc[0, :]
        #print(type(df_focus_2))
        df_focus_all = pd.concat([df_focus, df_focus_all], ignore_index=True)
        print(df_focus_all.info())
```
